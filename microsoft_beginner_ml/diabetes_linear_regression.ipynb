```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model, model_selection
```

Here I load the diabetes dataset but instead of loading it like the task I load it as a pandas dataframe so I can read what is going on.

```python
X,y = datasets.load_diabetes(return_X_y=True,as_frame = True)
print(X.shape)
print(y.shape)
```

We see there are a few columns Ill work with "column 3" as the task says which is BMI. 

s1 s2 s3 etc. are Sugar Level

Dataframe values seem to have a different meanings then I understand for example age is sometimes negative

```python
X
```

```python
y
```

Taking the BMI column and converting it to the same type as y. Then I reshape it to be 1 by number of original rows.

```python
X = X['bmi']
X = X.to_numpy()
X = X.reshape((-1,1))
```

```python
X
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size= 0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, linewidth = 4, color = 'green')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Effect of BMI on Disease Progression')
plt.show()
```

Task is done, Ill try some other stuff with the now

```python
print(model.score(X_test, y_test))
```

```python
X,y = datasets.load_diabetes(return_X_y=True, as_frame= True)
X = X['age'].to_numpy()
X_train, X_test, y_train, y_test = model_selection.train_test_split(X.reshape((-1,1)),y, test_size= 0.1)
model_3 = linear_model.LinearRegression()
model_3.fit(X_train,y_train)
y_pred = model_3.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, linewidth = 4, color = 'green')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Effect of BMI on Disease Progression')
plt.show()
```

```python
print(model_3.score(X_test, y_test))
```

Final results:

Assuming the dataframe columns are represented as what they mean (age is age of the person, sex is there sex, etc.), then the BMI is directly related to disease progression.

Weirdly, the age doesnt seem to contribute much if at all to the disease progression, this is likely due to the size of the dataset and more data would show a more clear relation.

Furthermore, using all the columns gives better results than just using age or BMI independantly to predict disease progression.
